```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("C://Users//bhara//Desktop//titanic_dataset.csv")
```

```python
data.head(20)
```

```python
data.isnull().mean()
```

```python
X_train,X_test,y_train,y_test=train_test_split(data[["Age","Fare"]],data["Survived"],test_size=0.5,random_state=0)
X_train.shape,X_test.shape
```

```python
X_train.isnull().mean()
```

```python
median=X_train.Age.median()
median
```

```python
mean=X_train.Age.mean()
mean=np.round(mean,1)
mean
```

```python
X_train["Age_median"]=X_train["Age"].fillna(median)
X_train["Age_mean"]=X_train["Age"].fillna(mean)
```

```python
X_train.tail(15)
```

```python
data.head(15)
```

```python
print("original variable variance:",X_train["Age"].var())
print("variance after mean imputation:",X_train["Age_mean"].var())
print("variance after median imputation",X_train["Age_median"].var())
```

```python
fig=plt.figure()
ax=fig.add_subplot(111)
X_train["Age"].plot(kind="kde",ax=ax)
X_train["Age_median"].plot(kind="kde",ax=ax,color="red")
X_train["Age_mean"].plot(kind="kde",ax=ax,color="green")
lines,labels=ax.get_legend_handles_labels()
ax.legend(lines,labels,loc="best")
```
