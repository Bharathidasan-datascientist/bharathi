```python
data.columns=["label","text"]
```

```python
# Preprocessing like tokenization,Removal punctuation,discarding stopwords using gensim
import gensim
```

```python
data["clean_data"]=data["text"].apply(lambda x:gensim.utils.simple_preprocess(x))
```

```python
data.head()
```

```python
#Split data into train and test
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(data['clean_data'],data["label"],test_size=0.2)
```

```python
# Train the word2vec model
```

```python
w2v_model=gensim.models.Word2Vec(X_train,window=5,min_count=2)
```

```python
#Explore the word vector for "king" based on our trained model
w2v_model.wv['king']
```

```python
#Find the most similar words to "king" based on word vectors from our trained model
```

```python
w2v_model.wv.most_similar('king')
```

```python
#Generate a list of words the Word2Vec model learned word vectors
```

```python
w2v_model.wv.key_to_index
```

```python
X_test.head(10)
```

```python
X_test.shape
import numpy as np
```

```python
#Generate aggregated sentence words based on the word vectors for each word in the sentence
w2v_vect=np.array([np.array([w2v_model.wv[i] for i in ls if i in w2v_model.wv.key_to_index])for ls in X_test])
```

```python
#Why is the length of the sentence different than the length of the sentence vector?
```

```python
for i,v in enumerate(w2v_vect):
    print(len(X_test.iloc[i]),len(v))
```

```python
#Computing sentence vectors by averaging the word vectors for the words contained in the sentence
```

```python
w2v_avg=[]
for vect in w2v_vect:
    if len(vect)!=0:
        w2v_avg.append(vect.mean(axis=0))
    else:
        w2v_avg.append(np.zeros(100))
```

```python
# are our sentence vector lengths consistence?
```

```python
for i,v in enumerate(w2v_avg):
    print(len(X_test.iloc[i]),len(v))
```
